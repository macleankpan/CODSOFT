### MACHINE LEARNING MODEL IMPLEMENTATION

Import libraries for Machine learning   

```python
from sklearn import preprocessing, linear_model
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
```

Label Encoder

```python
label_encoder = preprocessing.LabelEncoder()
read_data['species']= label_encoder.fit_transform(read_data['species'])

read_data['species'].unique()
```

From the output above, the value of this species have been converted to an array, 
- Iris-setosa = 0
- iris-versicolor = 1
- iris-virginica = 2 

```python
#to check if the values of species were converted to numeric form or not
read_data['species'].head()
```

Splitting the data into sets

```python
x= read_data.drop(['species'], axis=1)  #independent variable

y= read_data['species']                 #dependent variable
```

Declaring ration for traning and testing 

```python
x_sepal_train, x_sepal_test, y_species_train, y_species_test = train_test_split(x,y, random_state=0, test_size=30/100)
```

```python
x_sepal_train.shape, x_sepal_test.shape, y_species_train.shape, y_species_test.shape
```

```python
y_species_test
```

Linear Regression (Supervised Learning)
- It will be used to show the linear relationship between y(dependent variable) and x(independent variable)

```python
model=LinearRegression()
```

```python
#input data into the machine

model.fit(x_sepal_train, y_species_train)
y_species_predict = model.predict(x_sepal_test)
```

Check the accuracy of the linear regression model 

```python
score = model.score(x_sepal_test, y_species_test)
score = round(score * 100, 4)

print("Accuracy of model = ", str(score), " percent.")
```

```python
logistic_model = linear_model.LogisticRegression(max_iter=130)
logistic_model.fit(x_sepal_train, y_species_train)
```

```python
y_species_logistic_predict =logistic_model.predict(x_sepal_test)
```

```python
score_lg = logistic_model.score(x_sepal_test, y_species_test)
score_lg = round(score_lg * 100, 4)

print("Accuracy of model = ", str(score_lg), " percent.")
```

```python
cm = confusion_matrix(y_species_test, y_species_test)
cm
```

Plotting Confusion matrix using a heatmap

```python
sns.heatmap(cm, annot=True, cmap = "coolwarm")
plt.rcParams['figure.figsize']= (10,2)
plt.xlabel('PREDICTED VALUE')
plt.ylabel('SPECIES')
```

K-Nearest Neighbour MODEL

```python
knnModel = KNeighborsClassifier(n_neighbors=3)  #instantiate learning model (k =3)
knnModel.fit(x_sepal_train, y_species_train)
y_knn_pred = knnModel.predict(x_sepal_test)
```

Checking the Accuracy

```python
accuracy= accuracy_score(y_species_test, y_knn_pred)*100
accuracy
```

```python
print(classification_report(y_species_test, y_knn_pred))
```

```python
knn_cm = confusion_matrix(y_species_test, y_knn_pred)
knn_cm
```

```python
sns.heatmap(knn_cm, annot=True, cmap = "coolwarm")
plt.rcParams['figure.figsize']= (10,2)
plt.xlabel('PREDICTED VALUE')
plt.ylabel('SPECIES')
```

CHOOSING BEST MODEL FOR CLASSIFICATION

```python
scores_plt = [score, score_lg, accuracy]
algorithm = ['Linear Regression', 'Logistic Regression', "kNN"]
sns.set(rc={"figure.figsize": (11,6)})
plt.xlabel('Algorithms')
plt.ylabel(" Accuracy")

sns.barplot(x=algorithm, y=scores_plt)
plt.show()
```
